# Read Data

```python
import pandas as pd
```

# Pre-Processing

- Mengganti koma (,) dengan string kosong ('') pada seluruh nilai pada kolom Parking Duration (menghilangkan koma pada setiap data Parking Duration)
- Mengubah tipe data pada kolom Parking Duration menjadi integer (int), setelah nilai koma pada kolom tersebut dihapus.


```python
grafton_west["Parking Duration"] = grafton_west["Parking Duration"].replace(',','', regex=True).astype(int)
display(grafton_west)
```

```python
park_street["Parking Duration"] = park_street["Parking Duration"].replace(',','', regex=True).astype(int)
display(park_street)
```

```python
queen_annes["Parking Duration"] = queen_annes["Parking Duration"].replace(',','', regex=True).astype(int)
display(queen_annes)
```

Cek:
- "Entries" bernilai 1 jika dan hanya jika "Exits" bernilai 0
- "Parking Duration" bernilai tidak 0 ketika "Exits" bernilai 1
- "Parking Duration" bernilai 0 ketika "Exits" bernilai 0

```python
grafton_west['check'] = ((grafton_west['Parking Duration'] != 0) & (grafton_west['Exits'] == 1)) | ((grafton_west['Parking Duration'] == 0) & (grafton_west['Exits'] == 0) & (grafton_west['Entries'] == 1))
display(grafton_west)

num_error = grafton_west['check'].value_counts()[False]

# menampilkan jumlah nilai False
print("Jumlah data rusak pada dataset Grafton West:", num_error)
```

```python
park_street['check'] = ((park_street['Parking Duration'] != 0) & (park_street['Exits'] == 1)) | ((park_street['Parking Duration'] == 0) & (park_street['Exits'] == 0) & (park_street['Entries'] == 1))
display(park_street)

num_error = park_street['check'].value_counts()[False]

# menampilkan jumlah nilai False
print("Jumlah data rusak pada dataset Park Street:", num_error)
```

```python
queen_annes['check'] = ((queen_annes['Parking Duration'] != 0) & (queen_annes['Exits'] == 1)) | ((queen_annes['Parking Duration'] == 0) & (queen_annes['Exits'] == 0) & (queen_annes['Entries'] == 1))
display(queen_annes)

num_error = queen_annes['check'].value_counts()[False]

# menampilkan jumlah nilai False
print("Jumlah data rusak pada dataset Queen Annes:", num_error)
```

Menghapus data yang rusak

```python
grafton_west = grafton_west.drop(grafton_west[grafton_west['check'] == False].index)
grafton_west = grafton_west.drop(grafton_west[grafton_west['Exits'] == 0].index)
display(grafton_west)
```

```python
park_street = park_street.drop(park_street[park_street['check'] == False].index)
park_street = park_street.drop(park_street[park_street['Exits'] == 0].index)
display(park_street)
```

```python
queen_annes = queen_annes.drop(queen_annes[queen_annes['check'] == False].index)
queen_annes = queen_annes.drop(queen_annes[queen_annes['Exits'] == 0].index)
display(queen_annes)
```

'regex=True' pada metode replace() akan mengganti nilai secara literal, yaitu mengganti nilai yang persis sama dengan nilai yang diberikan pada parameter pertama dengan nilai pada parameter kedua.

# Inferensi Statistika

```python
import numpy as np
from scipy import stats
import seaborn as sns
from matplotlib import pyplot as plt
```

### Grafton West

```python
durasi_parkir_grafton_west = grafton_west["Parking Duration"].tolist()
```

```python
# memvisualisasikan distribusi data sampel

grafton_west_hist = sns.histplot(data=durasi_parkir_grafton_west, color = "blue", label = "Durasi Parkir", alpha=0.5, bins=80)

plt.legend()
plt.title("Histogram Grafton West")
plt.show()
```

```python
sns.boxplot(y=durasi_parkir_grafton_west)
plt.title("Boxplot Grafton West")
plt.grid(True)
```

#### Estimasi Rata-rata Populasi

```python
# menghitung statistik data sampel
n = len(durasi_parkir_grafton_west)
x_bar = np.mean(durasi_parkir_grafton_west)
s = stats.tstd(durasi_parkir_grafton_west) # method ini digunakan untuk stdev sampel

# menghitung estimasi 2 sisi interval rataan populasi berdasarkan rataan sampel
alpha = 0.05
# mencari nilai z_alpha_half sehingga P(Z>z_alpha_half) = 1 - alpha/2
z_alpha_half = stats.norm.ppf(1-alpha/2, loc=0, scale=1)

# menghitung standard error berdasarkan formula estimasi
standard_error = z_alpha_half * s / np.sqrt(n)

# menghitung batas bawah dan batas atas
batas_bawah = x_bar - standard_error
batas_atas = x_bar + standard_error

print("Estimasi 2 sisi rataan populasi")
print("-------------------------------")
print("Total data sampel\t: {}".format(n))
print("Rata-rata data sampel\t: {}".format(x_bar))
print("Stdev data sampel\t: {}".format(s))
print("\n")
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```

Dari hasil di atas, kita dapat menyimpulkan bahwa dengan tingkat kepercayaan 95%, rata-rata durasi parkir di Grafton West diperkirakan berada pada rentang antara 99.30 hingga 100.36 menit.

```python
# menghitung estimasi 1 sisi dari atas interval rataan populasi berdasarkan rataan sampel

alpha = 0.05
# mencari nilai z_alpha sehingga P(Z>z_alpha) = 1 - alpha
z_alpha = stats.norm.ppf(1-alpha, loc=0, scale=1)

# menghitung batas atas
batas_atas = x_bar + z_alpha * s / np.sqrt(n)
batas_bawah = x_bar - z_alpha * s / np.sqrt(n)

print("Estimasi 1 sisi dari atas rataan populasi")
print("-------------------------------")
print("Total data sampel\t: {}".format(n))
print("Rata-rata data sampel\t: {}".format(x_bar))
print("Stdev data sampel\t: {}".format(s))
print("\n")
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```

#### Estimasi Standar Deviasi Populasi

```python
# menghitung estimasi 2 sisi interval standar deviasi populasi berdasarkan rataan sampel

alpha = 0.05
df = n-1

# mencari nilai c_alpha_half sehingga P(Chi<c_alpha_half) = 1 - alpha/2
c_alpha_half = stats.chi2.ppf(1-alpha/2, df)
# mencari nilai c_alpha_half sehingga P(Chi<c_alpha_half) = alpha/2
c_one_min_alpha_half = stats.chi2.ppf(alpha/2, df)

# menghitung batas bawah dan batas atas
batas_bawah = s * np.sqrt(df/c_alpha_half)
batas_atas = s * np.sqrt(df/c_one_min_alpha_half)

print("Estimasi 2 sisi st dev populasi")
print("-------------------------------")
print("Total data sampel\t: {}".format(n))
print("Rata-rata data sampel\t: {}".format(x_bar))
print("Stdev data sampel\t: {}".format(s))
print("\n")
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```

Dari hasil di atas, kita dapat menyimpulkan bahwa dengan tingkat kepercayaan 95%, standar deviasi durasi parkir di Grafton West diperkirakan berada pada rentang antara 79.34 hingga 80.09 menit.

```python
bermasalah = len(grafton_west[grafton_west['Parking Duration'] > 480])
tidak_bermasalah = len(grafton_west[grafton_west['Parking Duration'] <= 480])
sampel = []
for i in range(bermasalah):
    sampel.append(1)
for i in range(tidak_bermasalah):
    sampel.append(0)
    
# menghitung statistik data sampel
n = len(sampel)
x_bar = np.mean(sampel)
s = stats.tstd(sampel) # method ini digunakan untuk stdev sampel

# menghitung estimasi 2 sisi interval rataan populasi berdasarkan rataan sampel

alpha = 0.05
df = n-1
# mencari nilai t_alpha_half sehingga P(T>t_alpha_half) = 1 - alpha/2
t_alpha_half = stats.t.ppf(1-alpha/2, df)

# menghitung standard error berdasarkan formula estimasi
standard_error = t_alpha_half * s / np.sqrt(n)

# menghitung batas bawah dan batas atas
batas_bawah = x_bar - standard_error
batas_atas = x_bar + standard_error

print("Estimasi 2 sisi rataan populasi")
print("-------------------------------")
print("Total data sampel\t: {}".format(n))
print("Rata-rata data sampel\t: {}".format(x_bar))
print("Stdev data sampel\t: {}".format(s))
print("\n")
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
print("\n")
print(f"Jadi, persentase banyaknya pengguna 'bermasalah' di Grafton West adalah {round(batas_bawah * 100,2)}% sampai dengan {round(batas_atas * 100,2)}% dengan tingkat kepercayaan sebesar {(1-alpha)*100}%")
```

### Skema Pembayaran Baru

```python
def new_scheme_calculate_fare(duration):
    if duration % 60 == 0:
        duration = (duration/60)+1
    else:
        duration = np.ceil(duration/60)

    if duration >= 0 and duration <= 1:
        return duration * 0.5 
    elif duration <= 2 and duration > 1:
        return (duration-1) * 0.75 + 0.5
    elif duration <= 8 and duration > 2:
        return (duration-2) * 1 + 1.25
    elif duration <= 24 and duration > 8 :
        return (duration-8) * 1.5 + 7.25
    elif duration > 24:
        return np.ceil(duration/24) * 31.25
```

```python
keseluruhan_data_baru = keseluruhan_data.assign(new_fare=keseluruhan_data['Parking Duration'].apply(new_scheme_calculate_fare))
display(keseluruhan_data_baru)
```

```python
pendapatan_keseluruhan_data_baru = keseluruhan_data_baru["new_fare"].tolist()

# menghitung statistik data sampel
n = len(pendapatan_keseluruhan_data_baru)
x_bar = np.mean(pendapatan_keseluruhan_data_baru)
s = stats.tstd(pendapatan_keseluruhan_data_baru) # method ini digunakan untuk stdev sampel

# menghitung estimasi 2 sisi interval rataan populasi berdasarkan rataan sampel
alpha = 0.05
# mencari nilai z_alpha_half sehingga P(Z>z_alpha_half) = 1 - alpha/2
z_alpha_half = stats.norm.ppf(1-alpha/2, loc=0, scale=1)

# menghitung standard error berdasarkan formula estimasi
standard_error = z_alpha_half * s / np.sqrt(n)

# menghitung batas bawah dan batas atas
batas_bawah = x_bar - standard_error
batas_atas = x_bar + standard_error

print("Estimasi 2 sisi rataan populasi")
print("-------------------------------")
print("Total data sampel\t: {}".format(n))
print("Rata-rata data sampel\t: {}".format(x_bar))
print("Stdev data sampel\t: {}".format(s))
print("\n")
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```
